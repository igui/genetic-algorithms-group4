```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
records = np.load('best_objectives_record-090316.npy', allow_pickle=True)
seeds, bestObjectives = records[()]['seeds'], records[()]['bestObjectives']

log = pd.read_csv('r0123456.csv', header=None, skiprows=2)
```

```python
# an informative plot
plt.boxplot(bestObjectives)
```

```python
begin = 200

f, ax = plt.subplots(2, 1, figsize=(13, 10), dpi=80)
ax[0].scatter(log.iloc[begin:, 1], log.iloc[begin:, 2], c='blue', s=0.05)
ax[0].plot(log.iloc[begin:, 1], log.iloc[begin:, 3], c='red')
ax[0].set_xlabel('second (s)')
ax[0].set_ylabel('Tour cost')
ax[0].set_title('Convergence plot of an experimental run')
ax[0].legend(['Mean tour cost', 'Best tour cost'])

ax[1].boxplot(bestObjectives, vert=False)
ax[1].set_xlabel('Tour cost')
ax[1].set_yticks([])
ax[1].set_title('Boxplot of best fitness (20 runs)')
```
